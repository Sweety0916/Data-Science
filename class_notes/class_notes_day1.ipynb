# **MATPLOTLIB LIBRARY **


```python

import matplotlib.pyplot as plt
```

1. Scatter plot with matplotlib

```python
plt.scatter(data['longitude'], data['latitude'])
plt.title('scatterplot')
plt.xlabel('longitude') 
plt.ylabel('latitude')
plt.show()
```

```python
plt.scatter(data['longitude'], data['latitude'],c=data['housing_median_age'],s=data['population'])
plt.title('scatterplot')
plt.xlabel('longitude') 
plt.ylabel('latitude')

plt.show()
```

```python
plt.scatter(data['longitude'], data['latitude'],c=data['housing_median_age'],s=data['population'])
plt.title('scatterplot')
plt.xlabel('longitude') 
plt.ylabel('latitude')
plt.colorbar()
plt.show()
```

2.  Barplot by matplotlib

```python
plt.bar(data['longitude'], data['latitude'])
plt.title('barplot')
plt.xlabel('longitude') 
plt.ylabel('latitude')
plt.show()
```

Histogram by matplotlib
it univariate in this we need only x-axis

```python
plt.hist(data['median_house_value'])
plt.show()
```

## **SEABORN**  all graph


1. scatter plot by seaborn
in seaborn scatterplot we use hue for color not c command

```python
import seaborn as sn

sn.scatterplot(x='median_house_value',y='latitude',data=data)
plt.show()
```

```python
import seaborn as sn

sn.scatterplot(x='median_house_value',y='latitude',data=data, hue='longitude')
plt.show()
```

2. lineplot plot by seaborn

```python
sn.lineplot(x='longitude',y='latitude',data=data)
plt.show()
```

```python
sn.lineplot(x='population',y='latitude',data=data, hue='longitude')
plt.show()
```

3. Histogram plot by seaborn

```python
sn.histplot(x='median_house_value',data=data, kde="true")
plt.show()
```

```python
sn.histplot(x='median_house_value',data=data, hue='longitude', kde="true")
plt.show()
```
